```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
import matplotlib
```

```python
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
```

```python
house_price=load_boston()
df=pd.DataFrame(house_price.data,columns=house_price.feature_names)
df['PRICE']=house_price.target
df.head()
```

```python
newX=df.drop('PRICE',axis=1)
newY=df['PRICE']
X_train,X_test,y_train,y_test=train_test_split(newX,newY,test_size=0.3,random_state=3)
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
coeff=lr.coef_
coeff
```

```python
plt.rcParams['figure.figsize']=(10,6)
plt.bar(X_train.columns,lr.coef_,linewidth=10)
plt.title("Regression Coefficients")
```

```python
X_test
```

```python
ridge_model=Ridge(alpha=10)
ridge_model.fit(X_train,y_train)
```

```python
ridge_model.coef_
```

```python
coeff_list=[[],[],[],[],[]] #RM,ZN,RAD,CRIM,TAX []
for i in range(1,201):
    rr=Ridge(alpha=i)
    rr.fit(X_train,y_train)
    #room
    coeff_list[0].append(rr.coef_[0]) #crim
    coeff_list[1].append(rr.coef_[5])# rm
    coeff_list[2].append(rr.coef_[1]) #zn
    coeff_list[3].append(rr.coef_[8])#rad
    coeff_list[4].append(rr.coef_[9])#tax

    
    
    
```

```python
plt.plot(coeff_list[0],label='CRIM')
plt.plot(coeff_list[1],label='RM')
plt.plot(coeff_list[2],label='ZN')
plt.plot(coeff_list[3],label='RAD')
plt.plot(coeff_list[4],label='TAX')
plt.ylabel('coefficients')
plt.xlabel('Value of Lambda')
plt.title("Ridge Regression")
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import Lasso
coeff_listl=[[],[],[],[],[]] #RM,ZN,RAD,CRIM,TAX []
for i in range(1,201):
    rr=Lasso(alpha=i)
    rr.fit(X_train,y_train)
    #room
    coeff_listl[0].append(rr.coef_[0]) #crim
    coeff_listl[1].append(rr.coef_[5])# rm
    coeff_listl[2].append(rr.coef_[1]) #zn
    coeff_listl[3].append(rr.coef_[8])#rad
    coeff_listl[4].append(rr.coef_[9])#tax

    
    
    
```

```python
plt.plot(coeff_listl[0],label='CRIM')
plt.plot(coeff_listl[1],label='RM')
plt.plot(coeff_listl[2],label='ZN')
plt.plot(coeff_listl[3],label='RAD')
plt.plot(coeff_listl[4],label='TAX')
plt.ylabel('coefficients')
plt.xlabel('Value of Lambda')
plt.title("Lasso Regression")
plt.legend()
plt.show()
```

```python
train_error_ridge=[]
train_error_lasso=[]
test_error_lasso=[]
test_error_ridge=[]
for i in [10,100,200]:
    #for ridge regression
    r_m=Ridge(alpha=i)
    r_m.fit(X_train,y_train)
    residuals_test=r_m.predict(X_test)-y_test
    test_error=np.mean(np.square(residuals_test))
    test_error_ridge.append(test_error)
    residuals_train=r_m.predict(X_train)-y_train
    train_error=np.mean(np.square(residuals_train))
    train_error_ridge.append(train_error)


for i in [10,100,200]:
    #for ridge regression
    l_m=Lasso(alpha=i)
    l_m.fit(X_train,y_train)
    residuals_test=l_m.predict(X_test)-y_test
    test_error=np.mean(np.square(residuals_test))
    test_error_lasso.append(test_error)
    residuals_train=l_m.predict(X_train)-y_train
    train_error=np.mean(np.square(residuals_train))
    train_error_lasso.append(train_error)
i=0
for j in[10,100,200]:
    print("Train error {} Test error {} for Ridge with alpha {} ".format(train_error_ridge[i],test_error_ridge[i],j))
    #print("Train error {} Test error {} for Lasso with alpha {} ".format(train_error_lasso[i],test_error_lasso[i],j))
    i=i+1

k=0
for j in[10,100,200]:
    print("Train error {} Test error {} for Ridge with alpha {} ".format(train_error_lasso[k],test_error_lasso[k],j))
    #print("Train error {} Test error {} for Lasso with alpha {} ".format(train_error_lasso[i],test_error_lasso[i],j))
    k=k+1
```
